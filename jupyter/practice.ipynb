```python
import pandas as pd
```

```python
df3 = pd.read_csv('./3. 2013년 이후 재무 데이터 (분기).csv')
df3
```

```python
# wide -> long format transformation

data_columns = df3.columns[3:]
data_columns
```

```python
date_columns = df3.columns[3:]  # 코드, 코드명, 아이템명 제외

df_long = df3.melt(
    id_vars=['코드', '코드명', '아이템명'],
    value_vars=date_columns,
    var_name='날짜',
    value_name='값'
)
df_long['날짜'] = pd.to_datetime(df_long['날짜'])
df_long['연도'] = df_long['날짜'].dt.year
df_long['값'] = pd.to_numeric(df_long['값'], errors='coerce')
```
