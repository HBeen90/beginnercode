```python
import pandas as pd
```

```python
df3 = pd.read_csv('./3. 2013년 이후 재무 데이터 (분기).csv')
df3
```

```python
# wide -> long format transformation

data_columns = df3.columns[3:]
data_columns
```

```python
date_columns = df3.columns[3:]  # 코드, 코드명, 아이템명 제외

df_long = df3.melt(
    id_vars=['코드', '코드명', '아이템명'],
    value_vars=date_columns,
    var_name='날짜',
    value_name='값'
)
df_long['날짜'] = pd.to_datetime(df_long['날짜'])
df_long['연도'] = df_long['날짜'].dt.year
df_long['값'] = pd.to_numeric(df_long['값'], errors='coerce')
```

```python
df_long
```

```python
# 필요한 항목만 추출 (자본총계, 액면가, 기말발행주식수)

target_items = ['자본총계(천원)', '액면가(원,US센트)', '기말발행주식수(보통주)(주)']
sub_df = df_long[df_long['아이템명'].isin(target_items)]
```

```python
# 각 연도의 '마지막 분기(연말)' 값만 선택

year_end = (
    sub_df.sort_values('날짜').groupby(['코드', '코드명', '아이템명', '연도']).last().reset_index()
)
```

```python
year_end
```

```python
# 아이템명을 컬럼으로 피벗 -> 새로운 테이블 짜기
pivot = year_end.pivot_table(index =['코드', '코드명', '연도'],
                             columns='아이템명',
                             values = '값').reset_index()
pivot.columns.name = None
pivot
```

```python
# 자본금 계산
# 자본총계 -> 천원 단위, 액면가 -> 원 단위
# 자본금 = 액면가(원) x 발행주식수(주)
# 차본금(천원) = 자본금(원) / 1000
pivot['자본금(천원)'] = pivot['액면가(원,US센트)'] * pivot['기말발행주식수(보통주)(주)'] /1000
pivot
```

```python
# 부실 의심 플래그
# 자본잠식률 50% 이상
pivot['자본잠식률(%)'] = (pivot['자본금(천원)']-pivot['자본총계(천원)'])/pivot['자본금(천원)'] * 100
pivot['관리종목의심(50%이상)'] = pivot['자본잠식률(%)'] >= 50
pivot['완전자본잠식'] = pivot['자본총계(천원)'] <= 0
pivot
```

```python
# ============================================
# 1. 연도 순으로 정렬 (rolling 계산 전 필수)
# ============================================
pivot = pivot.sort_values(['코드', '연도']).reset_index(drop=True)

# ============================================
# 2. 종목별로 "관리종목의심"이 2년 연속 True인지 체크
# ============================================
pivot['부실기업_2년연속'] = (
    pivot.groupby('코드')['관리종목의심(50%이상)']
    .transform(lambda x: x.rolling(window=2, min_periods=2).sum() == 2)
)

```

```python
company_status =(
    pivot.groupby(['코드', '코드명'])['부실기업_2년연속']
                   .any()
                   .reset_index()
                   .rename(columns = {'부실기업_2년연속' : '부실기업_여부'})
    )
```

```python
company_status = (
    pivot.groupby(['코드', '코드명'])['부실기업_2년연속']
    .any()  # 어느 연도에서든 한 번이라도 True면 그 종목은 부실기업으로 분류
    .reset_index()
    .rename(columns={'부실기업_2년연속': '부실기업_여부'})
)
```

```python
target_items = [
    '부채총계(천원)', '자본총계(천원)', '영업이익(천원)',
    '영업활동으로인한현금흐름(천원)', '당기순이익(천원)', '유동자산(천원)',
    '자산총계(천원)', '이익잉여금(천원)', '매출액(천원)',
    '시가총액(티커-상장예정주식수 포함)(백만원)'
]
```

```python

missing = [item for item in target_items if item not in df_long['아이템명'].unique()]
if missing:
    print(f"⚠️ STEP 2 - 데이터에 없는 아이템명: {missing}")
else:
    print("✅ STEP 2 - 모든 아이템명 확인 완료")
```

```python
# 분기별 재무 데이터 추출
sub_df = df_long[df_long['아이템명'].isin(target_items)].copy()
```

```python
quarter_pivot = sub_df.pivot_table(index = ['코드', '코드명', '날짜'],
                                   columns= '아이템명', values='값', aggfunc ='first').reset_index()
quarter_pivot
```

```python
quarter_pivot.columns.name = None
```

```python
quarter_pivot = quarter_pivot.sort_values(['코드', '코드명']).reset_index(drop= True)
```

```python
quarter_pivot
```

```python
#부채비율, 영업손실, 흑자도산, 유동성지표 계산
# 부채비율(%) = 부채총계(천원) / 자본총계(천원) *100
quarter_pivot['부채비율(%)'] = (quarter_pivot['부채총계(천원)']/quarter_pivot['자본총계(천원)']) * 100 
```

```python
# 유동비율(%) =  유동자산/유동부채 *100 -> 유동부채는 Dart 시스템에서 알수 있어
# 유동비율_추정 (%) = 유동자산 / 부채총계  * 100 

quarter_pivot['유동비율_추정(%)'] = quarter_pivot['유동자산(천원)']/quarter_pivot['부채총계(천원)'] * 100
```

```python
# 영업손실 -> 영업이익 <0
quarter_pivot['영업손실'] = quarter_pivot['영업이익(천원)'] <0
```

```python
# 흑자도산위험 = 영업활동으로인한현금흐름 <0 & 당기순이익 >0
quarter_pivot['흑자도산위험'] = (quarter_pivot['영업활동으로인한현금흐름(천원)'] < 0) & (quarter_pivot['당기순이익(천원)'] > 0) 
```

```python
# 부채과다
quarter_pivot['부채과다'] = quarter_pivot['부채비율(%)'] >= 200
# 유동성위험
quarter_pivot['유동성위험'] = quarter_pivot['유동비율_추정(%)'] < 50
```

```python
# 부실지표 연속성 체크
# 부채과다_4Qd연속
quarter_pivot['부채과다_4Q연속'] = (
    quarter_pivot.groupby('코드')['부채과다'].transform(lambda x : x.rolling(window=4, min_periods=4).sum()==4)
    )
```

```python
quarter_pivot['영업손실_4Q연속'] = (
    quarter_pivot.groupby('코드')['영업손실'].transform(lambda x : x.rolling(window=4, min_periods=4).sum()==4)
)
```

```python
quarter_pivot['흑자도산_2Q연속'] = quarter_pivot.groupby('코드')['흑자도산위험'].transform(lambda x : x.rolling(window=2, min_periods=2).sum()==2)
```

```python
quarter_pivot['유동성위험_2Q연속'] =(
    quarter_pivot.groupby('코드')['유동성위험'].transform(lambda x : x.rolling(window=2, min_periods=2).sum()==2)
)
```
